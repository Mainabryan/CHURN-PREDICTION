```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(y_true, y_pred,
                         labels=None,
                         title="Confusion Matrix - Customer Churn",
                         cmap="Blues",
                         figsize=(7, 5),
                         annot_size=14):
    """
    Plots a clean confusion matrix using seaborn heatmap.

    Parameters:
    -----------
    y_true : array-like
        True labels (0 = Stayed, 1 = Exited usually)
    y_pred : array-like
        Predicted labels
    labels : list of str, optional
        Names for the classes, e.g. ['Stayed', 'Exited']
    title : str
        Plot title
    cmap : str
        Colormap (Blues, Greens, Purples, etc.)
    figsize : tuple
        Figure size (width, height)
    annot_size : int
        Font size of numbers inside cells
    """
    # Default labels if none provided
    if labels is None:
        labels = ['Stayed (0)', 'Exited (1)']

    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)

    # Create figure
    plt.figure(figsize=figsize)

    # Heatmap
    sns.heatmap(
        cm,
        annot=True,
        fmt='d',
        cmap=cmap,
        xticklabels=labels,
        yticklabels=labels,
        annot_kws={"size": annot_size},
        cbar_kws={'label': 'Count'}
    )

    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.title(title, fontsize=14, pad=15)

    # Make sure labels are visible
    plt.xticks(rotation=0)
    plt.yticks(rotation=0)

    plt.tight_layout()
    plt.show()

    # Optional: also print the matrix numbers
    print("Confusion Matrix:")
    print(cm)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
import matplotlib.pyplot as plt

def plot_confusion_matrix_sklearn(y_true, y_pred,
                                 labels=None,
                                 title="Confusion Matrix",
                                 cmap='Blues',
                                 figsize=(7, 6)):
    if labels is None:
        labels = ['Stayed', 'Exited']

    cm = confusion_matrix(y_true, y_pred)

    disp = ConfusionMatrixDisplay(
        confusion_matrix=cm,
        display_labels=labels
    )

    fig, ax = plt.subplots(figsize=figsize)
    disp.plot(ax=ax, cmap=cmap, values_format='d')

    plt.title(title, fontsize=14, pad=15)
    plt.grid(False)
    plt.tight_layout()
    plt.show()

    print("Confusion Matrix:\n", cm)
```
